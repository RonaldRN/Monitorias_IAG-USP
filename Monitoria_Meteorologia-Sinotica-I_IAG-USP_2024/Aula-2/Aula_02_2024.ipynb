# Criando um DataFrame com o período completo (2007-2023) para a construção de um climograma e processamento dos ciclos diurno, diário, mensal e anual

**23. Selecionar os arquivos do ano 2007 até 2018**

```python
#Criar df com os dados do ano

dados_2007_2018 = pd.DataFrame()

for ano in range(2007,2019):

    # Lista de arquivos
    with ZipFile(str(ano)+".zip", "r") as lista:
        #print(*lista.namelist(), sep="\n")
        listaArquivos = lista.namelist()

    # Separar os arquivos de SC
    listaSC = []
    for i in range(0,len(listaArquivos)):
        if re.findall("A806", listaArquivos[i]):
            listaSC.append(listaArquivos[i])
            #print(listaArquivos[i])

    # Selecionar o arquivo na listaSC e indicar a linha do "header" ==> cabeçalho
    for i in listaSC:
        with ZipFile(str(ano)+".zip") as file:
            with file.open(i) as f:
                df = pd.read_csv(f, sep=";", header=8, encoding="ISO-8859-1", decimal=",")

            #Meta dados inicio do CSV, transpor para transformar em coluna
            with ZipFile(str(ano)+".zip") as file:
                with file.open(i) as f:
                    #h = pd.read_csv(f, sep=";", nrows=7, encoding="ISO-8859-1",decimal=",").set_index("REGIÃO:").T
                    h = pd.read_csv(f, sep=";", nrows=7, encoding="ISO-8859-1",decimal=",")

                    # Incluir os meta dados no final df
                    for i in range(0,len(h.columns)):
                        df[h.columns[i].replace(":","")] = h.iloc[0,i]

                    # Concatenas os arquivos
                    dados_2007_2018 = pd.concat([df, dados_2007_2018], ignore_index=True)

#renomear colunas
dados_2007_2018 = dados_2007_2018.rename(columns={'DATA (YYYY-MM-DD)':'Data','HORA (UTC)':'Hora UTC','RADIACAO GLOBAL (KJ/m²)':'RADIACAO GLOBAL (Kj/m²)','ESTAÇÃO':'ESTACAO', 'DATA DE FUNDAÇÃO (YYYY-MM-DD)':'DATA DE FUNDACAO'})

dados_2007_2018
```

# **Climograma**

O Climograma é um gráfico de representação das principais variáveis meteorológicas (precipitação e temperaturas) ao longo dos meses.

Por convenção da OMM, um climograma é criado a partir de médias de um período padronizado em 30 anos. Mas neste caso, o climograma vai ser gerado com a data disponível.

Para construir o nosso climograma, vamos mais uma vez usar o **groupby()** para agrupar os nossos dados de acordo com os meses **index.month** e, por fim, calcularmos as "normais" climatológicas.

**47. Fazer o agrupamento por mêses e fazer o cálculo da média para obter o valor climático**

```python
# FAZENDO UM CLIMOGRAMA

A701_climatologia = A701_media_diaria.groupby(A701_media_diaria.index.month).mean()
```

**48. Para o caso da precipitação é necessário fazer um pre-processamento desses dados:**
- Calcular os acumulados mensais para o período disponível.
- Calcular a média dos acumulados e agregar esse novo valor ao DataFrame (sustituir o valor anterior)

```python
# Calculando acumulados de precipitação mensais
# Fazer médias mensais dos acumulados de precipitação

# Acumulado mensal
A701_pp_mensal = A701_media_diaria['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'].groupby(pd.Grouper(freq = '1M')).sum()

# Média climatológica de precipitação
A701_climatologia['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'] = A701_pp_mensal.groupby(A701_pp_mensal.index.month).mean()
```

**49. Os climogramas apresentam dois eixos Y. O primeiro eixo é de precipitação e o segundo de temperatura, o que significa que todas as variáveis compartilham o mesmo eixo x dos meses.**

**O Matplotlib e Proplot podem criar eixos "gêmeos", tanto para o eixo x .twinx() quanto y .twiny():**

```python
# Fazendo um plot com dos eixos Y
fig, ax = plot.subplots(figsize = (8,3), tight = True)

# eixos gêmeos em X
ax2 = ax.twinx()

# Precipitação
ax.bar(plot.arange(1,12,1), A701_climatologia['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'], color = 'Blue', alpha = 0.5)

# Temperaturas
ax2.plot(plot.arange(1,12,1), A701_climatologia['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'], color = 'slate', label='INS')
ax2.plot(plot.arange(1,12,1), A701_climatologia['TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)'], color = 'bright red', label='MAX')
ax2.plot(plot.arange(1,12,1), A701_climatologia['TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)'], color = 'bright blue', label='MIN')

# Edição dos eixos
ax.format(title = 'Climograma - São Paulo - Mirante - A701',
          xlabel = 'Meses', ylabel = 'Precipitação (mm/mês)',
          xticklabels = ['JAN', 'FEV', 'MAR', 'ABR', 'MAI', 'JUN',\
                         'JUL', 'AUG', 'SET', 'OUT', 'NOV', 'DEZ'],
          xticks = plot.arange(1,12,1),
          xtickminor = False, ytickminor = False)

ax2.format(ylabel = 'Temperatura ($\degree$C)', ytickminor = False)

# Legenda
ax2.legend(bbox_to_anchor = (1.2, 0.6), ncols = 1, frameon = False)

plt.show()

# Salvando o plot
path = '/content/drive/MyDrive/USP-CURSOS/MONITORIA/Meteorologia-Sinotica-I/Aula-2'
# Salvando o plot
fig.savefig(path + '/Climograma-A701.png', format='png', dpi = 300)
```
